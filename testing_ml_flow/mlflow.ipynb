```python
import mlflow
import numpy as np
import pickle
from sklearn.metrics import accuracy_score, f1_score

mlflow.set_experiment("Evaluasi_Model_PKL")


X_test = np.load("X_test2.npy")
y_test = np.load("y_test2.npy")

if len(X_test.shape) > 2:
    X_test = X_test.reshape(X_test.shape[0], -1)
    
params = {
    "max_depth": 5,
    "learning_rate": 0.1,
    "n_estimators": 100,
    "random_state": 42
}

with open("model_stress2.pkl", "rb") as f:
    model = pickle.load(f)


with mlflow.start_run() as run:
    mlflow.log_params(params)
    mdel_info = mlflow.sklearn.log_model(
    model, 
    "model", 
    skops_trusted_types=["xgboost.core.Booster", "xgboost.sklearn.XGBClassifier"]
)

    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    mlflow.log_metric("accuracy", acc)
    mlflow.set_tag("Training info", "Evaluasi menggunakan data NPY dan model PKL")

    exp_id = run.info.experiment_id
   
df_runs = mlflow.search_runs(experiment_ids=[exp_id])
display(df_runs)

print("\nKolom yang tersedia di MLflow search_runs:")
print(df_runs.columns.tolist())

loaded_model = mlflow.sklearn.load_model(mdel_info.model_uri)
print("\nModel berhasil di-load dari MLflow:")
print(loaded_model)

prediction = loaded_model.predict(X_test)
f1 = f1_score(prediction, y_test, average='macro')

print(f"\nF1 Score dari loaded_model: {f1}")
```
